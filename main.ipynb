Importing Libraries
-- Here we will import all the necessary libraries to make our project

```python
import pandas as pd # to work with csv files
from sklearn.model_selection import train_test_split # to create train and test data
from sklearn.svm import SVC # main machine learning model - SUPPORT VECTOR MACHINE
from sklearn import metrics # To see the accuracy percentage of trained model
from sklearn.preprocessing import MinMaxScaler # to pre-process data
import numpy as np # to do scientific and mathematical calculations
```

Data Pre-Processing
-- Here we will pre-process the data so that we can easily use it in our model. Pre-processing is very important before using for training the model.

```python
data = pd.read_csv("Data/train.csv")
test_data = pd.read_csv("Data/test.csv")
submissions = pd.read_csv("sample_submission.csv")
```

Splitting Data Into Train And Test Data
-- Now we will split the data into testing and training. Even though you can use the test.csv for testing but don't have any label to evaluate the results that's why we are using spltting method for the data to successfully evaluate the model.

```python
features = data.iloc[0:,1:]
features
```

```python
labels = data["label"]
labels
```

```python
Xtrain , Xtest , Ytrain , Ytest = train_test_split(features , labels , test_size = 0.3 , random_state = 1)
```

```python
Xtest
```

```python
Ytest
```

Standorization Of Data
-- Now we will stadorize the data. This mean we will replace the real value of any cell (for ex. - 245) to the standord value which will be based on the all values of that column and will be assigned in some range (for ex. - 0.8 (range is 0 to 1)).

```python
norm = MinMaxScaler().fit(Xtrain)
Xtrain_norm = norm.transform(Xtrain)
X_test_norm = norm.transform(Xtest)
```

Training The Model
-- Here we will train the model. You can change the parameters like degree and C to see get different accuracy. Try to change it and then train your model.

```python
SVM  = SVC(kernel='poly' , degree=3 , C=1)
SVM.fit(Xtrain_norm, Ytrain)
```

```python
predict = SVM.predict(X_test_norm)
```

```python
x = metrics.accuracy_score(Ytest , predict)
print(x)
```

```python
predictor = SVM.predict(test_data)
print(predictor)
```

```python
#Creating CSV file for results
submissions['Label'] = predictor
submissions.to_csv("submission.csv",index=False)
```
